```python
import pandas as pd

# Load the CSV file into a DataFrame
train_df = pd.read_csv("bracket_training.csv")

# Display the first few rows to verify
train_df.head()
```

```python
# Assuming df is your DataFrame
null_counts = train_df.isnull().sum()

# Display NULL values in each column
print(null_counts)
```

```python
train_df = train_df.dropna(subset=['CustomerPostalCodeLatitude', 'CustomerPostalCodeLongitude', 'CustomerDMACode', 'CustomerDMADescription'])
```

```python
# Assuming df is your DataFrame
null_counts = train_df.isnull().sum()

# Display NULL values in each column
print(null_counts)
```

```python
institutes = pd.read_csv("Institutions.csv")
train_df = pd.merge(train_df, institutes[['InstitutionID', 'RegularSeasonAverageScore', 'RegularSeasonLosses', 'RegularSeasonWins']]
                  , left_on='RegionWinner_East', right_on='InstitutionID', how='left')

train_df.rename(columns={'RegularSeasonAverageScore': 'east_score', 'RegularSeasonLosses': 'east_losses', 'RegularSeasonWins': 'east_wins'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)


train_df.head()
```

```python
train_df = pd.merge(train_df, institutes[['InstitutionID', 'RegularSeasonAverageScore', 'RegularSeasonLosses', 'RegularSeasonWins']]
                  , left_on='RegionWinner_West', right_on='InstitutionID', how='left')

train_df.rename(columns={'RegularSeasonAverageScore': 'west_score', 'RegularSeasonLosses': 'west_losses', 'RegularSeasonWins': 'west_wins'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)


train_df.head()
```

```python
train_df['east_ratio'] = train_df['east_wins'] / (train_df['east_wins'] + train_df['east_losses'])
train_df['west_ratio'] = train_df['west_wins'] / (train_df['west_wins'] + train_df['west_losses'])

train_df.drop(['east_wins', 'west_wins' , 'east_losses', 'west_losses'], axis=1, inplace=True)

train_df.head()
```

```python
import numpy as np

train_df['east_west_rb'] = np.where((train_df['east_ratio'] > train_df['west_ratio'])
                                    & (train_df['east_score'] >= train_df['west_score']), train_df['RegionWinner_East'], train_df['RegionWinner_West'])

train_df.drop(['east_ratio', 'west_ratio' , 'east_score', 'west_score'], axis=1, inplace=True)

train_df.head()
```

```python
train_df = pd.merge(train_df, institutes[['InstitutionID', 'InstitutionConference']]
                  , left_on='RegionWinner_East', right_on='InstitutionID', how='left')
train_df.head()
```

```python
train_df.rename(columns={'InstitutionConference': 'east_conference'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)

train_df.head()
```

```python
train_df = pd.merge(train_df, institutes[['InstitutionID', 'InstitutionConference']]
                  , left_on='RegionWinner_West', right_on='InstitutionID', how='left')
train_df.rename(columns={'InstitutionConference': 'west_conference'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)

train_df.head()
```

```python
train_df = pd.merge(train_df, institutes[['InstitutionID', 'margin', 'fgp', 'ft']]
                  , left_on='RegionWinner_East', right_on='InstitutionID', how='left')

train_df.rename(columns={'margin': 'east_margin', 'fgp': 'east_fgp', 'ft': 'east_ft'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)

train_df = pd.merge(train_df, institutes[['InstitutionID', 'margin', 'fgp', 'ft']]
                  , left_on='RegionWinner_West', right_on='InstitutionID', how='left')

train_df.rename(columns={'margin': 'west_margin', 'fgp': 'west_fgp', 'ft': 'west_ft'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)


train_df.head()
```

```python
train_df = pd.merge(train_df, institutes[['InstitutionID', 'InstitutionDMACode']]
                  , left_on='RegionWinner_East', right_on='InstitutionID', how='left')

train_df.rename(columns={'InstitutionDMACode': 'east_dma'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)

train_df = pd.merge(train_df, institutes[['InstitutionID', 'InstitutionDMACode']]
                  , left_on='RegionWinner_West', right_on='InstitutionID', how='left')

train_df.rename(columns={'InstitutionDMACode': 'west_dma'}, inplace=True)

train_df.drop(['InstitutionID'], axis=1, inplace=True)

train_df['east_dma'] = np.where(train_df['east_dma'] == train_df['CustomerDMACode'], 'yes', 'no')
train_df['west_dma'] = np.where(train_df['west_dma'] == train_df['CustomerDMACode'], 'yes', 'no')


train_df.head()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import VarianceThreshold

# Define target and features
target_column = 'SemifinalWinner_East_West'
columns_to_exclude = ['SemifinalWinner_South_Midwest', 'NationalChampion', 
                      'RegionWinner_Midwest', 'RegionWinner_South', 
                      'CustomerDMADescription', 'CustomerAreaCode'
                     , 'CustomerPostalCode', 'NCAACustomerRecordCreated', 'BracketEntryCreatedDate', 'CustomerID'
                     , 'east_3p_total', 'west_3p_total', 'east_3p_success', 'west_3p_success', 'BracketEntryId'
                     , 'west_fgp', 'west_margin', 'west_ft', 'west_3p_perc']

# Selecting features (X) and target (y)
X_train = train_df.drop(columns=[target_column] + columns_to_exclude)
y_train = train_df[target_column]

# Identify categorical columns
categorical_cols = X_train.select_dtypes(include=['object']).columns

# Apply Ordinal Encoding to categorical features
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_train[categorical_cols] = encoder.fit_transform(X_train[categorical_cols])

# Splitting dataset
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

model = RandomForestClassifier(n_estimators=1000, max_depth=11, min_samples_split=3, min_samples_leaf=4, random_state=42)
model.fit(X_train, y_train)

# Predictions
# y_pred = model.predict(X_test)

# # Model evaluation
# accuracy = accuracy_score(y_test, y_pred)
# print(f"Model Accuracy: {accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Get feature importances
importances = model.feature_importances_

# Create a dataframe of feature importances
feature_names = X_train.columns  # Use the column names from X_train
feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})

# Sort the dataframe by importance
feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)

# Plot
plt.figure(figsize=(12, 8))
plt.bar(feature_importance_df['feature'], feature_importance_df['importance'])
plt.xticks(rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Random Forest Feature Importance')
plt.tight_layout()
plt.show()

# Print top 10 most important features
print("Top 10 Most Important Features:")
print(feature_importance_df.head(10))
```
